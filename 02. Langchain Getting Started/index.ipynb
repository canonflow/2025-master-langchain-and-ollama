```python
from dotenv import load_dotenv

load_dotenv("./../.env")
```

```python
import os
os.environ['LANGSMITH_ENDPOINT']
```

```python
from langchain_ollama import ChatOllama

base_url = os.getenv('OLLAMA_BASE_URL')
model_name = "llama3.2:latest"

llm = ChatOllama(
    base_url=base_url,  # Optional, default using localhost:11434
    model=model_name,  # Model Name
    temperature=0.7,  # Adjust creativity of the model, higher is more creative
    num_predict=1024  # How many tokens it is going to predict
)

llm_response = llm.invoke("How to make a bowl of cereal?")

print(llm_response.content)
```
